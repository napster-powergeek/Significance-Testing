```python
# MULTI FRAME LOGIC


try:
    import Tkinter as tk
except:
    import tkinter as tk
    
class SampleApp(tk.Tk):
    def __init__(self):
        tk.Tk.__init__(self)
        self._frame = None
        self.switch_frame(StartPage)

    def switch_frame(self, frame_class):
        new_frame = frame_class(self)
        if self._frame is not None:
            self._frame.destroy()
        self._frame = new_frame
        self._frame.pack()

class StartPage(tk.Frame):
    def __init__(self, master):
        tk.Frame.__init__(self, master)
        tk.Label(self, text="Start page", font=('Helvetica', 25, "bold")).pack(side="top", fill="x", padx=20, pady=20)
        tk.Button(self, text="Pre Test Evaluation",
                  command=lambda: master.switch_frame(PageOne)).pack(side="top", fill="x", padx=20, pady=20)
        tk.Button(self, text="Test Evaluation",
                  command=lambda: master.switch_frame(PageTwo)).pack(side="top", fill="x", padx=20, pady=20)

class PageOne(tk.Frame):
    def __init__(self, master):
        tk.Frame.__init__(self, master)
#         tk.Frame.configure(self,bg='blue')
        tk.Label(self, text="Page one", font=('Helvetica', 25, "bold")).pack(side="top", fill="x", padx=20, pady=20)
        tk.Button(self, text="Go back to start page",
                  command=lambda: master.switch_frame(StartPage)).pack(side="top", fill="x", padx=20, pady=20)

class PageTwo(tk.Frame):
    def __init__(self, master):
        tk.Frame.__init__(self, master)
#         tk.Frame.configure(self,bg='red')
        tk.Label(self, text="Page two", font=('Helvetica', 25, "bold")).pack(side="top", fill="x", padx= 20, pady=20)
        tk.Button(self, text="Go back to start page",
                  command=lambda: master.switch_frame(StartPage)).pack(side="top", fill="x", padx=20, pady=20)

if __name__ == "__main__":
    app = SampleApp()
    app.mainloop()
```
